## Python slice notation

The full syntax is sequence[start : stop : step]

### start

Index to begin at

Default: 0 (or end if step < 0)

### stop

Index to stop before

Default: end (or beginning if step < 0)

### step

Increment between items

Default: 1. Negative = reverse

### Works on lists too

nums = [10, 20, 30, 40, 50, 60] 

nums[1:4]      # → [20, 30, 40] 

nums[::2]      # → [10, 30, 50] 

nums[::-1]     # → [60, 50, 40, 30, 20, 10] 

nums[-3:]      # → [40, 50, 60] 

### Slices also work for assignment
nums[1:3] = [99, 88]

### nums is now [10, 99, 88, 40, 50, 60]

```python
nums = [10, 20, 30, 40, 50, 60]

print(nums[1:4])
print(nums[::2])
print(nums[-3:])
print(nums[::-1])
```

The syntax: s[start : stop : step]

All three parameters are optional and default to sensible values. Omitting one just uses its default.

### Rules to remember:

- stop is exclusive — the character at that index is not included
- Negative indices count from the end: -1 is the last character, -2 is second-to-last, etc.
- A negative step reverses the direction of traversal
- Out-of-bounds indices don't cause errors — Python clamps them silently
