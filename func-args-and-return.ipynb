```python
""" 
Variable length arguments: Positional Arguments
"""
def my_arg(*args):
    sum = 0
    for n in args:    
        sum +=n 
    return sum

print(my_arg(5, 5, 3, 2))
```

```python
def my_arg(*args):
    for n in args:
        print(n)

my_arg(1,5,9,"sandeep")
```

```python

""" 
Variable length arguments: Keyword Arguments
"""

def my_kwargs(n, **kwargs):    
    print(kwargs)
    n += kwargs["add"]
    n *= kwargs["multiply"]
    return n

ans = my_kwargs(5, add=2, multiply=3)

print(ans)
```

```python
def keyword_args(**kwargs):
    print(f"kwargs is {kwargs}")
    print(f"kwargs.items() is {kwargs.items()}")
    print("")

     # for loop requires tuples/sets so kwargs.items() returns tuples.
    for key, value in kwargs.items(): 
        print(f"Key is {key} and Value is {value}")

keyword_args(name="Siona", gender="Female")
```

```python
def positional_key_arguments(*args, **kwargs):
    for n in args:
        print(f"Positional argument is {n}")
    for key, value in kwargs.items():
        print(f"Key: {key} and Value: {value}")

""" 
If you switch *args with **kwargs, you will receive an error
"""

# def positional_key_arguments2(**kwargs, *args):
#     for n in args:
#         print(f"Positional argument is {n}")
#     for key, value in kwargs.items():
#         print(f"Key: {key} and Value: {value}")

positional_key_arguments(5,4,2,6, name="siona", gender="female")
```

```python
class Car:
    def __init__(self, **kwargs):
        self.make=kwargs.get("make")
        self.model=kwargs.get("model")
```

```python
myCar = Car(make="BMW", model="X7")

print(myCar.make)
print(myCar.model) 
```
